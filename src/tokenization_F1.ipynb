```python
# -*- coding: utf-8 -*-
"""NVTRS (2).ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1rnYj4r90WaDvuAcA8uIjfUT2pJWDHXlv
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# pytorch dataset
import torch # 파이토치라는 딥러닝 프레임워크 라이브러리입니다. 딥러닝 모델을 만들고 학습시키는 데 사용
from torch.utils.data import Dataset, DataLoader # 데이터를 모델에 넣기 좋게 관리하고, 미니배치(학습할 때 쓰는 데이터 묶음) 단위로 나눠서 불러오는 데 사용

# dataset split
from sklearn.model_selection import train_test_split

# model evaluate
from sklearn.metrics import precision_recall_fscore_support
# precision_score, recall_score, f1_score: 모델이 얼마나 잘 맞추는지 평가하는 데 사용

# model
import torch.nn as nn # 딥러닝 모델을 쉽게 만들 수 있도록 도와주는 모듈 (CNN 모델 만들 때 사용)
import torch.nn.functional as F # 신경망(NN)에서 자주 쓰는 함수(활성화 함수(activation func), 손실 함수(loss func 등)를 제공
# counter
from collections import Counter
import os

# progress bar
from tqdm import tqdm # 모델 학습 진행률 상황 시각화

# counter
from collections import Counter

# 1. 데이터 불러오기
df = pd.read_csv('/content/uniprotkb_AND_reviewed_true_2025_06_25.tsv', sep='\t')
df.head()

# 2. EC 번호 없는 결측치 제거
df = df[df['EC number'].notna()]

# 3. EC number 중 '-'와 'n'이 포함된 행 제거
df = df[~df['EC number'].str.contains('-', regex=False)]
df = df[~df['EC number'].str.contains('n', regex=False)]

df.head()

# 4. EC 번호 분리
def extract_2nd_level(ec_string):
    try:
        ec_list = ec_string.split(';') # [1.2.1.-; 1.2.1.44] 같은 여러개의 EC 번호를 ; 기준으로 나눔
        ec_trimmed = []
        for ec in ec_list:
            parts = ec.strip().split('.') # EC 번호 클래스 분리
            if len(parts) >= 2:
                trimmed = '.'.join(parts[:2]) # 2번째 클래스 까지만 저장
                ec_trimmed.append(trimmed)
        return ec_trimmed
    except:
        return []

df['ec_2nd_level'] = df['EC number'].apply(extract_2nd_level)

# 4. 클래스가 1로 시작하는 EC만 남기고 나머지 삭제
def keep_only_class_1(ec_list):
    return [ec for ec in ec_list if ec.startswith('1.')]

df['ec_2nd_level'] = df['ec_2nd_level'].apply(keep_only_class_1)
df = df[df['ec_2nd_level'].apply(len) > 0].reset_index(drop=True)

# 5. 각 고유 클래스(여기선 EC 2단계 번호)에 숫자 인덱스를 붙여주는 작업
# 클래스 multilabel vector 만드는 것

def encode_multilabel_2nd_level(ec_list):
    vec = np.zeros(len(ec2_to_idx), dtype=int)
    for ec in ec_list:
        if ec in ec2_to_idx:
            vec[ec2_to_idx[ec]] = 1
    return vec

num_multilabel_samples = df[df['ec_2nd_level'].apply(lambda x: len(x) >= 2)].shape[0]
print(f"✅ 라벨이 2개 이상인 다중 라벨 샘플 개수: {num_multilabel_samples}")

# 1. 시퀀스 데이터 전처리
AMINO_ACIDS = 'ACDEFGHIKLMNPQRSTVWYX'
INVALID_AMINO_ACIDS = 'BZOUJ'

# 유효한 아미노산 문자 목록
VALID_CHARS = set(AMINO_ACIDS + INVALID_AMINO_ACIDS)

def is_valid_sequence(seq):
    return all(ch.upper() in VALID_CHARS for ch in seq)

before = len(df)
df = df[df['Sequence'].apply(is_valid_sequence)].reset_index(drop=True)
after = len(df)
print(f"✅ 유효 문자만 포함된 시퀀스 필터링: {before} → {after}")

def clean_sequence(seq):
    seq = seq.upper()
    for ch in INVALID_AMINO_ACIDS:
        seq = seq.replace(ch, 'X')
    return seq

# 2. 시퀀스 클렌징
df['Cleaned_Sequence'] = df['Sequence'].apply(clean_sequence)

# 3. 중복 제거
df = df.drop_duplicates(subset='Cleaned_Sequence').reset_index(drop=True)

# 6. 고유 EC 추출 및 인덱스 매핑
all_ecs_2nd = sorted(set([ec for ecs in df['ec_2nd_level'] for ec in ecs]))
ec2_to_idx = {ec: i for i, ec in enumerate(all_ecs_2nd)}
idx_to_ec2 = {i: ec for ec, i in ec2_to_idx.items()}

print(f"✅ 최종 EC 클래스 수: {len(ec2_to_idx)}")

from collections import defaultdict

# 1. Cleaned_Sequence 기준으로 EC 리스트 및 원본 EC number 병합
sequence_to_ecs = defaultdict(list)
sequence_to_raw_ec = defaultdict(list)

for _, row in df.iterrows():
    seq = row['Cleaned_Sequence']
    ecs = row['ec_2nd_level']
    raw_ec = row['EC number']
    sequence_to_ecs[seq].extend(ecs)
    sequence_to_raw_ec[seq].append(raw_ec)

# 2. 중복 제거 및 정렬
merged_data = []
for seq in sequence_to_ecs:
    unique_ecs = sorted(set(sequence_to_ecs[seq]))
    merged_raw_ec = ';'.join(sorted(set(sequence_to_raw_ec[seq])))
    merged_data.append({
        'Cleaned_Sequence': seq,
        'ec_2nd_level': unique_ecs,
        'EC number': merged_raw_ec
    })

# 3. 다른 메타 정보 가져오기 (Cleaned_Sequence 기준)
meta_cols = ['Entry', 'Entry Name', 'Sequence']  # 필요한 컬럼 추가 가능
df_meta = df.drop_duplicates(subset='Cleaned_Sequence')[['Cleaned_Sequence'] + meta_cols]

# 4. 병합
merged_df = pd.DataFrame(merged_data)
df = pd.merge(merged_df, df_meta, on='Cleaned_Sequence', how='left')

# 6. label_vector 생성
df['label_vector'] = df['ec_2nd_level'].apply(encode_multilabel_2nd_level)

# 7. 컬럼 순서 정리 (ec_2nd_level 다음에 label_vector 배치)
df = df[['Entry', 'Entry Name', 'EC number', 'Sequence', 'ec_2nd_level', 'label_vector', 'Cleaned_Sequence']]

# 5. 희소 클래스(10개 이하) 제거
all_ecs_flat = [ec for ecs in df['ec_2nd_level'] for ec in ecs]
ec_counts = Counter(all_ecs_flat)
rare_ecs = set([ec for ec, cnt in ec_counts.items() if cnt <= 10])

def has_rare_ec(ec_list):
    return any(ec in rare_ecs for ec in ec_list)

df = df[~df['ec_2nd_level'].apply(has_rare_ec)].reset_index(drop=True)

# 전체 EC를 펼쳐서 카운팅
all_ec2_flat = [ec for sublist in df['ec_2nd_level'] for ec in sublist]
ec2_counts = Counter(all_ec2_flat)

# EC 번호 오름차순으로 정렬하여 출력
print("✅ EC 2단계 클래스별 개수 (오름차순 정렬):")
for ec in sorted(ec2_counts.keys(), key=lambda x: list(map(float, x.split('.')))):
    print(f"{ec}: {ec2_counts[ec]}")

ec2_all_list = [ec for sublist in df['ec_2nd_level'] for ec in sublist]

unique_ec2_list = sorted(set(ec2_all_list))
print(unique_ec2_list)
print(len(unique_ec2_list))

num_multilabel_samples = df[df['ec_2nd_level'].apply(lambda x: len(x) >= 2)].shape[0]
print(f"✅ 라벨이 2개 이상인 다중 라벨 샘플 개수: {num_multilabel_samples}")

print(len(df))

# K-mer 방식

K = 3  # 사용할 K-mer 길이

# 1. K-mer 추출 함수
def generate_kmers(seq, k=K):
    seq = seq.upper().strip()
    return [seq[i:i+k] for i in range(len(seq) - k + 1)]

# 2. 전체 데이터에서 K-mer 모으기
all_kmers = Counter()
for seq in df['Cleaned_Sequence']:
    all_kmers.update(generate_kmers(seq))

# 상위 20개 K-mer와 등장 횟수 출력
print(all_kmers.most_common(50))

# all_kmers: Counter 객체라고 가정
all_unique_kmers = list(all_kmers.keys())  # 전체 K-mer 리스트 추출

# 인덱스 부여 (PAD=0, UNK=1 이후 2부터 시작)
all_unique_kmers = [kmer for kmer in all_unique_kmers if kmer not in ('PAD', 'UNK')]
kmer_to_idx = {kmer: idx + 2 for idx, kmer in enumerate(all_unique_kmers)}
kmer_to_idx['PAD'] = 0
kmer_to_idx['UNK'] = 1

# 전체 vocab size는 다음과 같이 계산 가능
VOCAB_SIZE = len(kmer_to_idx)
print(f"✅ 전체 사용된 K-mer 개수 (VOCAB_SIZE): {VOCAB_SIZE}")

# 5. kmer 시퀀스 인코딩
def encode_kmer_sequence(seq, max_len=512, k=K):
    kmers = generate_kmers(seq, k)
    encoded = []

    for kmer in kmers:
        if kmer in kmer_to_idx:
            encoded.append(kmer_to_idx[kmer])
        else:
            encoded.append(kmer_to_idx['UNK'])

    # 자르거나 PAD로 채우기
    if len(encoded) > max_len:
        encoded = encoded[:max_len]
    else:
        encoded += [kmer_to_idx['PAD']] * (max_len - len(encoded))

    return encoded

# 시퀀스 인코딩 적용
df['encoded_kmers'] = df['Cleaned_Sequence'].apply(lambda x: encode_kmer_sequence(x, max_len=512))

from itertools import chain

# ec_2nd_level 컬럼이 리스트 형태라고 가정하고 모든 값 평탄화
all_ecs = list(chain.from_iterable(df['ec_2nd_level']))

# 고유한 EC 번호 추출
unique_ecs = set(all_ecs)

print(f"💡 고유 EC 번호 개수: {len(unique_ecs)}")
print(f"📋 EC 번호 목록: {sorted(unique_ecs)}")

class ProteinDataset(Dataset):

    # 데이터를 메모리에 올려두고, 인덱스로 쉽게 접근할 수 있게 준비
    def __init__(self, dataframe): # 초기와 함수
        self.sequences = dataframe['encoded_kmers'].tolist() # 숫자로 변환된 단백질 서열
        self.labels = dataframe['label_vector'].tolist() # 멀티라벨 벡터

    # 데이터셋의 크기를 파이토치가 알 수 있게 함
    def __len__(self):
        return len(self.sequences) # 전체 데이터(서열)의 개수를 반환

    def __getitem__(self, idx):
        x = torch.LongTensor(self.sequences[idx])   # 서열(정수 시퀀스)을 LongTensor(정수형 텐서)로 변환
        y = torch.FloatTensor(self.labels[idx])     # 다중(multi-hot) 라벨을 FloatTensor(실수형 텐서, 멀티라벨용)로 변환
        return x, y

# 파이토치로 데이터셋을 정리하는 이유
# - PyTorch의 DataLoader를 사용하면 모델 학습 시 자동으로 데이터를 배치 단위로 불러옴
# - 라벨을 텐서(Tensor)로 변환하면 GPU에서도 빠르게 계산한 가능



# 첫 번째 샘플을 통한 데이터 셋 체크
dataset = ProteinDataset(df)


x, y = dataset[0]
print("Input shape:", x.shape)
print("Label shape:", y.shape)
print("Sample X:", x[:10])  # 앞 10개 시퀀스 토큰
print("Sample Y:", y.nonzero())  # 라벨에 1인 위치만 보기

dataset

df

from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split

X = df['encoded_kmers'].tolist()
y = df['label_vector'].tolist()
X_tensor = pad_sequence([torch.tensor(x, dtype=torch.long) for x in X], batch_first=True)
y_tensor = torch.tensor(y, dtype=torch.float32)

X_train, X_val, y_train, y_val = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

dataloaders = {
    'train': DataLoader(TensorDataset(X_train, y_train), batch_size=32, shuffle=True),
    'val': DataLoader(TensorDataset(X_val, y_val), batch_size=32, shuffle=False)
}

import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision.datasets as dset
import torchvision.transforms as T
from torch.utils.data import DataLoader
from torch.utils.data import sampler
import matplotlib.pyplot as plt
import os
import copy

#!pip install einops
from einops import rearrange, repeat, reduce
from einops.layers.torch import Rearrange


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
dtype = torch.long

class TokPosEmbedding(nn.Module):
    def __init__(self, num_kmers, d_model, dropout_ratio, max_len=512):
        super().__init__()
        self.tokEmbedding = nn.Embedding(num_kmers, d_model, padding_idx=0)
        self.cls_token = nn.Parameter(torch.randn(1, 1, d_model))
        self.posEmbedding = nn.Embedding(max_len + 1, d_model)
        self.d_model = d_model
        self.dropout = nn.Dropout(dropout_ratio)
    def forward(self, src):
        x = self.tokEmbedding(src)           # (batch, seq_len, d_model)
        b = x.shape[0]
        cls_tokens = self.cls_token.expand(b, -1, -1)  # (batch, 1, d_model)
        x = torch.cat([cls_tokens, x], dim=1)          # (batch, seq_len+1, d_model)
        batch_size = x.shape[0]
        src_len = x.shape[1]
        pos = torch.arange(0, src_len, dtype=torch.long, device=x.device)
        pos = pos.unsqueeze(0).expand(batch_size, src_len)
        x = self.dropout(x * np.sqrt(self.d_model) + self.posEmbedding(pos))
        return x

class MultiHeadAttentionLayer(nn.Module):
    def __init__(self, d_model, nhead, dropout_ratio):
        super().__init__()
        self.d_model = d_model
        self.nhead = nhead
        self.head_dim = d_model // nhead
        self.qLinear = nn.Linear(d_model, d_model)
        self.kLinear = nn.Linear(d_model, d_model)
        self.vLinear = nn.Linear(d_model, d_model)
        self.oLinear = nn.Linear(d_model, d_model)
        self.dropout = nn.Dropout(dropout_ratio)

    def forward(self, query, key, value, mask=None):  # mask 인자 추가
        Q = self.qLinear(query)
        K = self.kLinear(key)
        V = self.vLinear(value)
        Q = rearrange(Q, 'b l (h d) -> b h l d', h=self.nhead)
        K = rearrange(K, 'b l (h d) -> b h l d', h=self.nhead)
        V = rearrange(V, 'b l (h d) -> b h l d', h=self.nhead)
        weight = torch.matmul(Q, rearrange(K, 'b h l d -> b h d l')) / np.sqrt(self.head_dim)
        if mask is not None:
            # mask shape: [batch, seq_len] -> [batch, 1, 1, key_len]
            mask = mask.unsqueeze(1).unsqueeze(2)   # (B,1,1,L)
            weight = weight.masked_fill(mask == 0, float('-inf'))
        attention = torch.softmax(weight, dim=-1)
        dropped_attention = self.dropout(attention)
        c = torch.matmul(dropped_attention, V)
        c = rearrange(c, 'b h l d -> b l (h d)')
        output = self.oLinear(c)
        return output, attention

class PositionWiseFeedForwardLayer(nn.Module):
    def __init__(self, d_model, ff_dim, dropout_ratio):
        super().__init__()
        self.linear1 = nn.Linear(d_model, ff_dim)
        self.linear2 = nn.Linear(ff_dim, d_model)
        self.dropout = nn.Dropout(dropout_ratio)

    def forward(self, x):
        x = self.dropout(nn.functional.gelu(self.linear1(x)))
        x = self.linear2(x)
        return x

class EncoderLayer(nn.Module):
    def __init__(self, d_model, nhead, ff_dim, dropout_ratio):
        super().__init__()
        self.layerNorm1 = nn.LayerNorm(d_model)
        self.layerNorm2 = nn.LayerNorm(d_model)
        self.multiHeadAttentionLayer = MultiHeadAttentionLayer(d_model, nhead, dropout_ratio)
        self.positionWiseFeedForward = PositionWiseFeedForwardLayer(d_model, ff_dim, dropout_ratio)
        self.dropout = nn.Dropout(dropout_ratio)

    def forward(self, src, mask=None):
        _src = self.layerNorm1(src)
        _src, attention = self.multiHeadAttentionLayer(_src, _src, _src, mask)
        src = src + self.dropout(_src)
        _src = self.layerNorm2(src)
        _src = self.positionWiseFeedForward(_src)
        src = src + self.dropout(_src)
        return src, attention

class Encoder(nn.Module):
    def __init__(self, d_model, n_layers, nhead, ff_dim, dropout_ratio):
        super().__init__()
        self.layers = nn.ModuleList([EncoderLayer(d_model, nhead, ff_dim, dropout_ratio) for _ in range(n_layers)])

    def forward(self, src, mask=None):
        attentions = []
        for layer in self.layers:
            src, attention = layer(src, mask)
            attentions.append(attention)
        return src, attentions

class Transformer(nn.Module):
    def __init__(self, num_kmers, d_model, n_layers, nhead, ff_dim, dropout_ratio, output_dim, max_len=512):
        super().__init__()
        self.encEmbedding = TokPosEmbedding(num_kmers, d_model, dropout_ratio, max_len)
        self.encoder = Encoder(d_model, n_layers, nhead, ff_dim, dropout_ratio)
        self.layerNorm = nn.LayerNorm(d_model)
        self.linear = nn.Linear(d_model, output_dim)

    def forward(self, src, mask=None):
        src = self.encEmbedding(src)  # (batch, seq_len+1, d_model)
        # mask도 [CLS] 토큰 포함해줘야 함
        if mask is not None:
            mask = torch.cat([torch.ones((mask.size(0), 1), dtype=torch.bool, device=mask.device), mask], dim=1)
        enc_src, attentions = self.encoder(src, mask)
        cls_token = enc_src[:, 0, :]
        cls_token = self.layerNorm(cls_token)
        output = self.linear(cls_token)
        return output, attentions

from sklearn.metrics import f1_score

def trainer_sam(model_name, model, criterion, optimizer, num_epochs, dataloaders):
    model.to(device)
    best_model_wts = copy.deepcopy(model.state_dict())

    train_losses, val_losses = [], []
    train_f1_macros, val_f1_macros = [], []
    train_f1_micros, val_f1_micros = [], []

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}\n----------')

        # ---- Training ----
        model.train()
        running_loss, total_samples = 0.0, 0
        all_train_labels, all_train_preds = [], []

        for inputs, labels in dataloaders['train']:
            inputs = inputs.to(device)
            labels = labels.to(device)
            mask = (inputs != 0)
            outputs, _ = model(inputs, mask=mask)
            probs = torch.sigmoid(outputs)
            preds = (probs > 0.35).int()

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.first_step(zero_grad=True)

            outputs_perturbed, _ = model(inputs, mask=mask)
            loss_perturbed = criterion(outputs_perturbed, labels)
            loss_perturbed.backward()
            optimizer.second_step(zero_grad=True)

            batch_size = inputs.size(0)
            running_loss += loss.item() * batch_size
            total_samples += batch_size

            all_train_labels.append(labels.cpu().numpy())
            all_train_preds.append(preds.cpu().numpy())

        epoch_train_loss = running_loss / total_samples
        train_losses.append(epoch_train_loss)

        all_train_labels = np.concatenate(all_train_labels, axis=0)
        all_train_preds = np.concatenate(all_train_preds, axis=0)
        f1_train_macro = f1_score(all_train_labels, all_train_preds, average='macro', zero_division=0)
        f1_train_micro = f1_score(all_train_labels, all_train_preds, average='micro', zero_division=0)
        train_f1_macros.append(f1_train_macro)
        train_f1_micros.append(f1_train_micro)

        print(f'Train Loss: {epoch_train_loss:.4f} | Macro F1: {f1_train_macro:.4f} | Micro F1: {f1_train_micro:.4f}')

        # ---- Validation ----
        model.eval()
        val_loss, val_total = 0.0, 0
        all_val_labels, all_val_preds = [], []
        with torch.no_grad():
            for inputs, labels in dataloaders['val']:
                inputs = inputs.to(device)
                labels = labels.to(device)
                mask = (inputs != 0)
                outputs, _ = model(inputs, mask=mask)
                probs = torch.sigmoid(outputs)
                preds = (probs > 0.35).int()
                loss = criterion(outputs, labels)
                batch_size = inputs.size(0)
                val_loss += loss.item() * batch_size
                val_total += batch_size

                all_val_labels.append(labels.cpu().numpy())
                all_val_preds.append(preds.cpu().numpy())

        epoch_val_loss = val_loss / val_total
        val_losses.append(epoch_val_loss)
        all_val_labels = np.concatenate(all_val_labels, axis=0)
        all_val_preds = np.concatenate(all_val_preds, axis=0)
        f1_val_macro = f1_score(all_val_labels, all_val_preds, average='macro', zero_division=0)
        f1_val_micro = f1_score(all_val_labels, all_val_preds, average='micro', zero_division=0)
        val_f1_macros.append(f1_val_macro)
        val_f1_micros.append(f1_val_micro)

        print(f'Val Loss: {epoch_val_loss:.4f} | Macro F1: {f1_val_macro:.4f} | Micro F1: {f1_val_micro:.4f}')

    return model, train_losses, val_losses, train_f1_macros, val_f1_macros, train_f1_micros, val_f1_micros


def trainer(model_name, model, criterion, optimizer, num_epochs, dataloaders):
    model.to(device)
    best_model_wts = copy.deepcopy(model.state_dict())

    train_losses, val_losses = [], []
    train_f1_macros, val_f1_macros = [], []
    train_f1_micros, val_f1_micros = [], []

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}\n----------')

        # ---- Training ----
        model.train()
        running_loss, total_samples = 0.0, 0
        all_train_labels, all_train_preds = [], []

        for inputs, labels in dataloaders['train']:
            inputs = inputs.to(device)
            labels = labels.to(device)
            mask = (inputs != 0)
            outputs, _ = model(inputs, mask=mask)
            probs = torch.sigmoid(outputs)
            preds = (probs > 0.35).int()

            loss = criterion(outputs, labels)
            optimizer.zero_grad()            # [1] optimizer grad 초기화
            loss.backward()                  # [2] 역전파
            optimizer.step()                 # [3] 파라미터 업데이트

            batch_size = inputs.size(0)
            running_loss += loss.item() * batch_size
            total_samples += batch_size

            all_train_labels.append(labels.cpu().numpy())
            all_train_preds.append(preds.cpu().numpy())

        epoch_train_loss = running_loss / total_samples
        train_losses.append(epoch_train_loss)

        all_train_labels = np.concatenate(all_train_labels, axis=0)
        all_train_preds = np.concatenate(all_train_preds, axis=0)
        f1_train_macro = f1_score(all_train_labels, all_train_preds, average='macro', zero_division=0)
        f1_train_micro = f1_score(all_train_labels, all_train_preds, average='micro', zero_division=0)
        train_f1_macros.append(f1_train_macro)
        train_f1_micros.append(f1_train_micro)

        print(f'Train Loss: {epoch_train_loss:.4f} | Macro F1: {f1_train_macro:.4f} | Micro F1: {f1_train_micro:.4f}')

        # ---- Validation ----
        model.eval()
        val_loss, val_total = 0.0, 0
        all_val_labels, all_val_preds = [], []
        with torch.no_grad():
            for inputs, labels in dataloaders['val']:
                inputs = inputs.to(device)
                labels = labels.to(device)
                mask = (inputs != 0)
                outputs, _ = model(inputs, mask=mask)
                probs = torch.sigmoid(outputs)
                preds = (probs > 0.35).int()
                loss = criterion(outputs, labels)
                batch_size = inputs.size(0)
                val_loss += loss.item() * batch_size
                val_total += batch_size

                all_val_labels.append(labels.cpu().numpy())
                all_val_preds.append(preds.cpu().numpy())

        epoch_val_loss = val_loss / val_total
        val_losses.append(epoch_val_loss)
        all_val_labels = np.concatenate(all_val_labels, axis=0)
        all_val_preds = np.concatenate(all_val_preds, axis=0)
        f1_val_macro = f1_score(all_val_labels, all_val_preds, average='macro', zero_division=0)
        f1_val_micro = f1_score(all_val_labels, all_val_preds, average='micro', zero_division=0)
        val_f1_macros.append(f1_val_macro)
        val_f1_micros.append(f1_val_micro)

        print(f'Val Loss: {epoch_val_loss:.4f} | Macro F1: {f1_val_macro:.4f} | Micro F1: {f1_val_micro:.4f}')
        scheduler.step()
        # -- epoch별 validation 마지막에서 --
        np.save("val_labels.npy", all_val_labels)
        np.save("val_preds.npy", all_val_preds)
        print('중간 F1:', f1_val_macro, f1_val_micro)

    return model, train_losses, val_losses, train_f1_macros, val_f1_macros, train_f1_micros, val_f1_micros



class SAM(torch.optim.Optimizer):
    def __init__(self, params, base_optimizer, rho=0.05, **kwargs):
        assert rho >= 0.0#, f"Invalid rho, should be non-negative: {rho}"

        defaults = dict(rho=rho, **kwargs)
        super(SAM, self).__init__(params, defaults)

        self.base_optimizer = base_optimizer(self.param_groups, **kwargs)
        self.param_groups = self.base_optimizer.param_groups

    # (1) Maximization for the perturbed loss
    @torch.no_grad()
    def first_step(self, zero_grad=False):
        grad_norm = self._grad_norm()
        for group in self.param_groups:
            scale = group["rho"] / (grad_norm + 1e-12) #1e-12는 분모 0안되게

            for p in group["params"]:
                if p.grad is None:
                    continue
                self.state[p]["old_p"] = p.data.clone()
                e_w = p.grad * scale.to(p)
                p.add_(e_w)  # climb to the local maximum "w + e(w)"

        if zero_grad: self.zero_grad()

    # (2) Minimization of perturbed loss
    @torch.no_grad()
    def second_step(self, zero_grad=False):
        for group in self.param_groups:
            for p in group["params"]:
                if p.grad is None:
                    continue
                p.data = self.state[p]["old_p"]  # get back to "w" from "w + e(w)"

        self.base_optimizer.step()  # do the actual "sharpness-aware" update

        if zero_grad: self.zero_grad()

    @torch.no_grad()
    def step(self, closure=None):
        assert closure is not None, "Sharpness Aware Minimization requires closure, but it was not provided"
        closure = torch.enable_grad()(closure)  # the closure should do a full forward-backward pass

        self.first_step(zero_grad=True)
        closure()
        self.second_step()

    def _grad_norm(self):
        shared_device = self.param_groups[0]["params"][0].device  # put everything on the same device, in case of model parallelism
        norm = torch.norm(
                    torch.stack([
                        p.grad.norm(p=2).to(shared_device)
                        for group in self.param_groups for p in group["params"]
                        if p.grad is not None
                    ]),
                    p=2
               )
        return norm

len(kmer_to_idx)

num_kmers = len(kmer_to_idx)    # 예시 (실제 값에 맞게!)
d_model = 128
n_layers = 2
nhead = 4
ff_dim = 256
dropout_ratio = 0.2
output_dim = 23
max_len = 512
num_epochs = 50

model = Transformer(
    num_kmers,
    d_model,
    n_layers,
    nhead,
    ff_dim,
    dropout_ratio,
    output_dim,
    max_len
)
model=model.to(device)
#criterion = nn.BCEWithLogitsLoss()
pos_weights = (y_tensor.shape[0] - y_tensor.sum(dim=0)) / (y_tensor.sum(dim=0) + 1e-6)
pos_weights = torch.tensor(pos_weights, dtype=torch.float32)
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weights.to(device))
from torch.optim.lr_scheduler import StepLR

optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)
scheduler = StepLR(optimizer, step_size=100, gamma=0.5)

# (1) 라벨 매핑 값이 train/val/전체에서 모두 같은지 체크
print("ec2_to_idx:", ec2_to_idx)
print("unique_ecs:", sorted(ec2_to_idx.keys()))

# 1. 시퀀스와 라벨벡터 맵핑
seq2label = dict(zip(df['Cleaned_Sequence'], df['label_vector']))

# 2. Validation set에서 임의로 5개 샘플 추출 (혹은 전체에서)
sample_idx = np.random.choice(len(X_val), size=5, replace=False)
for i in sample_idx:
    # dataloader용 index가 pandas index와 다르면, X_val[i]가 df index와 직접 매칭 안됨
    # 그러니 X_val, y_val의 i번째를 추출
    seq_tensor = X_val[i]              # shape: (512,)
    label_tensor = y_val[i]            # shape: (output_dim,)
    # DF에서 원본 시퀀스/라벨벡터 찾기 (padded_seq로 검색)
    arr = seq_tensor.cpu().numpy()
    # padded_seq가 numpy array이기 때문에, DF에서 똑같이 찾을 수 있음
    match = df[df['encoded_kmers'].apply(lambda x: np.array_equal(x, arr))]
    print("-------")
    if not match.empty:
        s = match.iloc[0]
        print("시퀀스:", s['Cleaned_Sequence'][:50], "...")
        print("원본 label_vector:", s['label_vector'])
        print("Val용 label_tensor:", label_tensor.cpu().numpy())
        print("ec_2nd_level:", s['ec_2nd_level'])
    else:
        print("매칭 실패! (padded_seq 일치 없음)")

# 라벨 shape, dtype, 순서 체크
print("train X shape:", X_train.shape)
print("train y shape:", y_train.shape)
print("val X shape:", X_val.shape)
print("val y shape:", y_val.shape)

print("전체 X_tensor shape:", X_tensor.shape)
print("전체 y_tensor shape:", y_tensor.shape)

# 인덱스와 EC번호 연결 확인
for idx, ec in enumerate(sorted(ec2_to_idx, key=lambda k: ec2_to_idx[k])):
    print(f"index={idx}, EC={ec}")

# 평가용 라벨, 원본 라벨 완전 일치 확인
# val dataloader에서 5개만 비교
from itertools import islice
for i, (inputs, labels) in enumerate(islice(dataloaders['val'], 5)):
    print("batch_y[0]:", labels[0].cpu().numpy())
    # batch_x[0]와 동일한 padded_seq 찾기
    arr = inputs[0].cpu().numpy()
    match = df[df['encoded_kmers'].apply(lambda x: np.array_equal(x, arr))]
    if not match.empty:
        print("원본 label_vector:", match.iloc[0]['label_vector'])
    else:
        print("매칭 실패!")
    print("-----")




model, train_losses, val_losses, train_f1_macros, val_f1_macros, train_f1_micros, val_f1_micros = trainer('TF_1', model, criterion, optimizer, num_epochs, dataloaders)


import matplotlib.pyplot as plt

plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

from sklearn.metrics import f1_score
import numpy as np

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in dataloaders['val']:  # 검증 셋
        inputs = inputs.to(device)
        labels = labels.cpu().numpy()
        mask = (inputs != 0)
        outputs, _ = model(inputs, mask=mask)
        probs = torch.sigmoid(outputs)
        preds = (probs > 0.35).cpu().numpy().astype(int)
        all_preds.append(preds)
        all_labels.append(labels)

all_preds = np.concatenate(all_preds, axis=0)
all_labels = np.concatenate(all_labels, axis=0)

f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
f1_micro = f1_score(all_labels, all_preds, average='micro', zero_division=0)

print(f"Macro F1-score: {f1_macro:.4f}")
print(f"Micro F1-score: {f1_micro:.4f}")

plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

from sklearn.metrics import f1_score
import numpy as np

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in dataloaders['val']:  # 검증 셋
        inputs = inputs.to(device)
        labels = labels.cpu().numpy()
        mask = (inputs != 0)
        outputs, _ = model(inputs, mask=mask)
        probs = torch.sigmoid(outputs)
        preds = (probs > 0.35).cpu().numpy().astype(int)
        all_preds.append(preds)
        all_labels.append(labels)

all_preds = np.concatenate(all_preds, axis=0)
all_labels = np.concatenate(all_labels, axis=0)

f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
f1_micro = f1_score(all_labels, all_preds, average='micro', zero_division=0)

val_labels2 = np.load("val_labels.npy")
val_preds2 = np.load("val_preds.npy")
print('최종 F1 재계산:', f1_score(val_labels2, val_preds2, average='macro'), f1_score(val_labels2, val_preds2, average='micro'))

print(f"Macro F1-score: {f1_macro:.4f}")
print(f"Micro F1-score: {f1_micro:.4f}")

# 1. (모든 배치의 확률값, 라벨 저장)
probs_list = []
labels_list = []

with torch.no_grad():
    for inputs, labels in dataloaders['val']:
        inputs = inputs.to(device)
        mask = (inputs!=0)
        outputs, _ = model(inputs, mask=mask)
        probs = torch.sigmoid(outputs)
        probs_list.append(probs.cpu().numpy())
        labels_list.append(labels.cpu().numpy())

# 2. 합치기
all_probs = np.concatenate(probs_list, axis=0)   # (전체 샘플 수, 클래스 수)
all_labels = np.concatenate(labels_list, axis=0) # (전체 샘플 수, 클래스 수)

# 3. threshold별 f1-score 계산
thresholds = np.arange(0.05, 0.96, 0.05)
f1_macros, f1_micros = [], []

for thresh in thresholds:
    all_preds = (all_probs > thresh).astype(int)
    f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    f1_micro = f1_score(all_labels, all_preds, average='micro', zero_division=0)
    f1_macros.append(f1_macro)
    f1_micros.append(f1_micro)
    print(f"Threshold={thresh:.2f} | Macro F1: {f1_macro:.4f} | Micro F1: {f1_micro:.4f}")



# 최고 F1-score일 때의 threshold 출력
best_macro_idx = np.argmax(f1_macros)
best_micro_idx = np.argmax(f1_micros)
print(f"\n[Best Macro F1] Threshold={thresholds[best_macro_idx]:.2f}, F1={f1_macros[best_macro_idx]:.4f}")
print(f"[Best Micro F1] Threshold={thresholds[best_micro_idx]:.2f}, F1={f1_micros[best_micro_idx]:.4f}")

# 그래프 그리기
plt.plot(thresholds, f1_macros, label='Macro F1')
plt.plot(thresholds, f1_micros, label='Micro F1')
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs Threshold')
plt.legend()
plt.show()

# 저장된 npy 파일 불러오기
labels = np.load('val_labels.npy')
preds = np.load('val_preds.npy')

print("val_labels shape:", labels.shape)
print("val_preds shape:", preds.shape)

# 값 일부 출력 (앞 5개만)
print("val_labels 샘플:")
print(labels[:5])

print("val_preds 샘플:")
print(preds[:5])

# 값이 0, 1만인지 체크 (예상과 다르면 threshold 이후인지, sigmoid 통과 안한 값인지 확인!)
print("라벨 unique 값:", np.unique(labels))
print("예측값 unique 값:", np.unique(preds))
```
